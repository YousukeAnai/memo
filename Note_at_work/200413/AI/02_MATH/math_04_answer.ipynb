Exercise_1  

以下のエントロピーを求めよ。（単位はビットとする。）  
ヒント：$log_23 = 1.58$

①出る目に偏りのないサイコロをふったとき、出た目を表す確率変数Xのエントロピーを求めよ。  

②1と6が$\frac{1}{3}$の確率で出現し、その他の目は$\frac{1}{12}$で出るサイコロをふったとき、出た目を表す確率変数Xのエントロピーを求めよ。  

Answer  

①  

$\displaystyle H(X) = - \sum_{i=1}^{6} p^{(i)} log_2p^{(i)}$  

　　　$\displaystyle = -6・\frac{1}{6}log_2\frac{1}{6}$  
   
　　　$\displaystyle = -log_2\frac{1}{6}$  

　　　$\displaystyle = log_26$  

　　　$\displaystyle = log_22・3$  

　　　$\displaystyle = log_22 + log_23$  

　　　$\displaystyle = 1 + 1.58$  

　　　$\displaystyle = 2.58$  
   

②  

$\displaystyle H(X) = - \sum_{i=1}^{6} p^{(i)} log_2p^{(i)}$  

　　　$\displaystyle = -2・\frac{1}{3}log_2\frac{1}{3} -4・\frac{1}{12}log_2\frac{1}{12}$  
   
　　　$\displaystyle = \frac{2}{3}log_23 +\frac{1}{3}log_22^2・3$  
   
　　　$\displaystyle = \frac{2}{3}・1.58 +\frac{1}{3}(2 + 1.58)$  
   
　　　$\displaystyle \fallingdotseq  2.25$  


Exercise_2  

公平なコインと以下の確率分布に従うサイコロがあるとする。  

$\displaystyle P(1) = \frac{1}{2}, P(2) = \frac{1}{4}, P(3) = \frac{1}{8}, P(4) = \frac{1}{16}, P(5) = \frac{1}{32}, P(6) = \frac{1}{32}$

コインを振って出た表と裏を表す確率変数を$X$とし、サイコロの出た目を表す確率変数を$Y$とする。  

①エントロピー$H(X)$を求めよ。  

②エントロピー$H(Y)$を求めよ。  

③エントロピー$H(X, Y)$を求めよ。  

※単位はビットとする

　　


Answer  

①  

$\displaystyle H(X) =  - \frac{1}{2}log_2\frac{1}{2} - \frac{1}{2}log_2\frac{1}{2} $  

　　　 $\displaystyle =  \frac{1}{2} +  \frac{1}{2} = 1 $  


②  

$\displaystyle H(Y) =  - \frac{1}{2}log_2\frac{1}{2} - \frac{1}{4}log_2\frac{1}{4} - \frac{1}{8}log_2\frac{1}{8} - \frac{1}{16}log_2\frac{1}{16} - 2・\frac{1}{32}log_2\frac{1}{32}$  

　　　 $\displaystyle =  \frac{1}{2} +  \frac{1}{2} +  \frac{3}{8} +  \frac{1}{4} + \frac{5}{16} $  

　　　 $\displaystyle =  \frac{31}{16} = 1.9375$  
    
③　　  

X,Yの同時確率は以下のようになる。   

$\displaystyle P(表, 1) =  P(裏, 1) = \frac{1}{4} $  

$\displaystyle P(表, 2) =  P(裏, 2) = \frac{1}{8} $  

$\displaystyle P(表, 3) =  P(裏, 3) = \frac{1}{16} $  

$\displaystyle P(表, 4) =  P(裏, 4) = \frac{1}{32} $  

$\displaystyle P(表, 5) =  P(裏, 5) = \frac{1}{64} $  

$\displaystyle P(表, 6) =  P(裏, 6) = \frac{1}{64} $  

よって、  

$\displaystyle H(X, Y) = -2・ \frac{1}{4}log_2\frac{1}{4} -2・ \frac{1}{8}log_2\frac{1}{8}-2・ \frac{1}{16}log_2\frac{1}{16} -2・ \frac{1}{32}log_2\frac{1}{32} -4・ \frac{1}{64}log_2\frac{1}{64}$

　　　 　$\displaystyle =  1 +  \frac{3}{4} +  \frac{1}{2} +  \frac{5}{16} + \frac{3}{8} $  

　　　 　$\displaystyle =  \frac{47}{16} = 2.9375 $

公平なコインAと以下の確率分布に従うコインBとCがあるとする。  

$\displaystyle P_B(表) = \frac{1}{4}, P_B(裏) = \frac{3}{4}, P_C(表) = \frac{1}{8}, P_C(裏) = \frac{7}{8}$  

$\displaystyle D_{KL}(P_A || P_B)$と$\displaystyle D_{KL}(P_A || P_C)$はどちらが大きいか。
